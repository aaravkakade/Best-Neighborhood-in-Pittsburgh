```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("employment-status-for-the-population-16-years-and-over.csv")

data.head()

data['employment_rate'] = data['Estimate; In labor force:']/ data['Estimate; Total:']

data.plot(x='Neighborhood', y='employment_rate', kind='bar')

plt.rcParams["figure.figsize"] = (50, 10)

plt.title("Employment Rate Over Pittsburgh Neighborhoods")
plt.show()
```

# Conclusion

Most of the neighborhoods have a solid employment rate around 50 percent or higher. There are some outliers like Glen Hazel, Lower Lawrencville, and a few other neighborhoods. The top 3 neighborhoods are North Shore, Strip District, and South Shore. Based on my knowledge from living in Pittsburgh, these neighborhoods are very active and have a lot of activity and food places.
