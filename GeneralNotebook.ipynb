# Introduction

The first thing we did when we started working on this project was look through the datasets on the WPRDC website and find ones related to certain aspects of neighborhoods. We wanted to create a solid, data-driven reason for our chosen best neighborhood and that meant having solid data. There were multiple metrics we found that could have supplied enough data and would be easy to explore like safety and access to health. The metric we settled on was community involvement and how active people were in their neighborhoods.

# The Metric

We chose community involvement as a metric because, with lots of involvement and contribution, communities are more socially connected. This increases socialization and social opportunities like block parties, trick-or-treating, and other community events. After we decided on this metric, we broke it down into 3 sub-metrics for each of us to measure: eco-friendliness, amount of parks, and employment rate. 

## Eco-Friendliness



```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'ecodata.csv'
data = pd.read_csv(file_path)

data['Eco_Friendly_Score'] = (
    data['Commute to Work: Public Transportation (2010)'] +
    data['Commute to Work: Walk (2010)'] -
    data['Commute to Work: Drive Alone (2010)']
)


most_eco_friendly = data.loc[data['Eco_Friendly_Score'].idxmax()]


plt.figure(figsize=(20, 10))
plt.bar(data['Neighborhood'], data['Eco_Friendly_Score'], color='green')
plt.title('Eco-Friendliness Scores by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Eco-Friendly Score')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## Park Counts

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("pittsburgh-parks.csv")


plt.rcParams["figure.figsize"] = (20, 10)

data=data["neighborhood"].value_counts()
data.plot.bar()
plt.show()
```

## Employment Rate

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("employment-status-for-the-population-16-years-and-over.csv")

data.head()

data['employment_rate'] = data['Estimate; In labor force:']/ data['Estimate; Total:']

data.plot(x='Neighborhood', y='employment_rate', kind='bar')

plt.rcParams["figure.figsize"] = (50, 10)

plt.title("Employment Rate Over Pittsburgh Neighborhoods")
plt.show()
```

# Best Neighborhood

To find the best neighborhood, we tried to find a neighborhood that performed well in each sub metric that we analyzed. This looked different when looking at each sub metric based on the variability of the dataset like the employment rate and eco friendliness data had a lot of variability but the parks did not have that much. Based on all this we were looking for a neighborhood with an employment rate above at least 60 percent, an eco friendly score of around -0.65, and have at least 4 or 5 parks. We found that Hazlewood performed well in each category and was the best neighborhood from all 3 of these categories.
